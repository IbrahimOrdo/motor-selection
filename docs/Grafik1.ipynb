```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the simulated data from CSV
df = pd.read_csv(r'C:\Users\IbrahimOrdo\OneDrive\Masaüstü\motor-selection\simulated_user_data.csv')

# Show the first few rows to check the data
print(df.head())

ages = df['Age']
heights = df['Height']
weights = df['Weight']

# Grafik
plt.scatter(heights, weights, color='blue')
plt.title('Boy ve Kilo Dağılımı')
plt.xlabel('Boy (cm)')
plt.ylabel('Kilo (kg)')
plt.show()
```

```python
# Örnek veriler: Haftalık Seyahat Mesafeleri (km)
names = df['Name']
travel_distances = df['Weekly Travel Distance']

# Grafik
plt.bar(names, travel_distances, color='green')

# Başlık ve etiketler
plt.title('Haftalık Seyahat Mesafesi')
plt.xlabel('Kullanıcılar')
plt.ylabel('Seyahat Mesafesi (km)')

# Grafik gösterimi
plt.show()
```

```python

# 1. Monthly Income vs Age (Bar Chart)
def plot_income_vs_expenses(incomes, ages, labels):
    plt.bar(labels, incomes, color='blue', alpha=0.6, label='Income')
    plt.bar(labels, ages, color='red', alpha=0.6, label='Age')
    plt.title('Monthly Income vs Expenses')
    plt.xlabel('Users')
    plt.ylabel('Amount (USD)')
    plt.legend()
    plt.show()

# Example Data
incomes = df['Annual Income']
ages = df['Age']

# Plotting the charts
plot_income_vs_expenses(incomes, ages, names)
```
