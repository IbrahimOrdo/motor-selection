```python
import matplotlib.pyplot as plt
import numpy as np

# Örnek veriler
heights = [180, 170, 160, 175, 185, 172, 168, 178, 190, 165]
weights = [80, 65, 70, 85, 90, 60, 68, 75, 95, 62]

# Grafik
plt.scatter(heights, weights, color='blue')
plt.title('Boy ve Kilo Dağılımı')
plt.xlabel('Boy (cm)')
plt.ylabel('Kilo (kg)')
plt.show()
```

```python
# Örnek veriler: Haftalık Seyahat Mesafeleri (km)
names = ['Kullanıcı 1', 'Kullanıcı 2', 'Kullanıcı 3', 'Kullanıcı 4', 'Kullanıcı 5']
travel_distances = [50.5, 35.0, 60.0, 40.0, 55.0]

# Grafik
plt.bar(names, travel_distances, color='green')

# Başlık ve etiketler
plt.title('Haftalık Seyahat Mesafesi')
plt.xlabel('Kullanıcılar')
plt.ylabel('Seyahat Mesafesi (km)')

# Grafik gösterimi
plt.show()
```

```python

# 1. Monthly Income vs Expenses (Bar Chart)
def plot_income_vs_expenses(incomes, expenses, labels):
    plt.bar(labels, incomes, color='blue', alpha=0.6, label='Income')
    plt.bar(labels, expenses, color='red', alpha=0.6, label='Expenses')
    plt.title('Monthly Income vs Expenses')
    plt.xlabel('Users')
    plt.ylabel('Amount (USD)')
    plt.legend()
    plt.show()

# 2. Age vs Driving Experience (Line Chart)
def plot_age_vs_experience(ages, experiences, labels):
    plt.plot(labels, ages, marker='o', color='green', label='Age')
    plt.plot(labels, experiences, marker='x', color='orange', label='Driving Experience')
    plt.title('Age vs Driving Experience')
    plt.xlabel('Users')
    plt.ylabel('Years')
    plt.legend()
    plt.show()

# Example Data
users = ['User 1', 'User 2', 'User 3', 'User 4', 'User 5']
incomes = [3000, 4000, 3500, 5000, 4500]
expenses = [2500, 3000, 2800, 4500, 4000]
ages = [25, 30, 28, 35, 40]
experiences = [5, 10, 6, 12, 15]

# Plotting the charts
plot_income_vs_expenses(incomes, expenses, users)
plot_age_vs_experience(ages, experiences, users)
```
